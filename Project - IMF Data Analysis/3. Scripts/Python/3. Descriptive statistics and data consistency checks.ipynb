## Flagging countries with outlier values

```python
# checking column names
df_overall_cpi_pcps_merged.columns
```

```python
# identifying lower and upper margins for outliers in 'consumer_price_index_all_items'

overall_cpi_q1 = df_overall_cpi_pcps_merged['consumer_price_index_all_items'].quantile(0.25)
overall_cpi_q3 = df_overall_cpi_pcps_merged['consumer_price_index_all_items'].quantile(0.75)
overall_cpi_iqr = overall_cpi_q3 - overall_cpi_q1

overall_cpi_lower_margin = overall_cpi_q1 - 1.5*overall_cpi_iqr
overall_cpi_upper_margin = overall_cpi_q3 + 1.5*overall_cpi_iqr

# table
margins_overall_cpi = [['overall_cpi_lower_margin', overall_cpi_lower_margin], ['overall_cpi_upper_margin', overall_cpi_upper_margin]]
overall_cpi_outlier_margins = pd.DataFrame(data = margins_overall_cpi, columns = ['margins', 'value'])
overall_cpi_outlier_margins
```

```python
# creating a list of countries with outlier values
outlier_cuntries_overall_cpi = df_overall_cpi_pcps_merged[(df_overall_cpi_pcps_merged['consumer_price_index_all_items'] < overall_cpi_lower_margin) + (df_overall_cpi_pcps_merged['consumer_price_index_all_items'] > overall_cpi_upper_margin)]['country_name'].unique().tolist()
outlier_cuntries_overall_cpi
```

```python
# creating a list of outlier country flags
outlier_country_cpi = []
 
for value in df_overall_cpi_pcps_merged['country_name'] :
    if value in outlier_cuntries_overall_cpi :
        outlier_country_cpi.append('Outlier country')
    else :
        outlier_country_cpi.append('Non-outlier country')

# creating flag list for cpi elements and pcps merged
outlier_country_cpi_elements = []
 
for value in df_cpi_elements_pcps_merged['country_name'] :
    if value in outlier_cuntries_overall_cpi:
        outlier_country_cpi_elements.append('Outlier country')
    else :
        outlier_country_cpi_elements.append('Non-outlier country')
```

```python
# assigning outlier country flag list as values of outlier_country_flag columns
df_overall_cpi_pcps_merged['outlier_country_flag'] = outlier_country_cpi 
df_cpi_elements_pcps_merged['outlier_country_flag'] = outlier_country_cpi_elements
```

```python
# checking values of the outlier flag columns
df_overall_cpi_pcps_merged['outlier_country_flag'].value_counts(dropna = False)
```

```python
# checking values of the outlier flag columns
df_cpi_elements_pcps_merged['outlier_country_flag'].value_counts(dropna = False)
```

## Checking for columns with mixed data types

```python
# checking for variables with mixed datatypes in overall cpi and pcps merged dataframe
for col in df_overall_cpi_pcps_merged.columns.tolist():
  weird = (df_overall_cpi_pcps_merged[[col]].applymap(type) != df_overall_cpi_pcps_merged[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_overall_cpi_pcps_merged[weird]) > 0:
    print (col)
```

```python
# checking for variables with mixed datatypes in cpi elements and pcps merged dataframe
for col in df_cpi_elements_pcps_merged.columns.tolist():
  weird = (df_cpi_elements_pcps_merged[[col]].applymap(type) != df_cpi_elements_pcps_merged[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cpi_elements_pcps_merged[weird]) > 0:
    print (col)
```
