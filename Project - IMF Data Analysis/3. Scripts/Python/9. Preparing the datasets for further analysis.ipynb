## Assigning dates to the dataframes

### Assigning date to overall cpi and pcps dataframe

```python
# creating date lists for overall cpi and pcps merged dataframe

date_cpi_pcps = []
for value in df_cpi_pcps['month_number']:
    index = np.array(list(map(int, df_cpi_pcps[df_cpi_pcps['month_number'] == value].index.values)))
    year = np.array(list(map(int, df_cpi_pcps.loc[index, 'year'])))
    next_month = np.array(list(map(int, df_cpi_pcps.loc[index, 'month_number'] + 1)))
    
    if value in list(range(1, 12, 1)):
        date_cpi_pcps.append((datetime.date(year[0], next_month[0], 1) - datetime.timedelta(days = 1)).isoformat())
    else: date_cpi_pcps.append(datetime.date(year[0], 12, 31).isoformat())
```

```python
# checking the list
date_cpi_pcps
```

```python
# creating new date column on the 'df_cpi_pcps' dataframe
df_cpi_pcps['date'] = date_cpi_pcps
df_cpi_pcps.head(288)
```

### Assigning date to cpi elements and pcps merged dataframe

```python
# creating date lists for cpi elements and pcps merged dataframe

date_cpi_elements_pcps = []
for value in df_cpi_elements_pcps['month_number']:
    index = np.array(list(map(int, df_cpi_elements_pcps[df_cpi_elements_pcps['month_number'] == value].index.values)))
    year = np.array(list(map(int, df_cpi_elements_pcps.loc[index, 'year'])))
    next_month = np.array(list(map(int, df_cpi_elements_pcps.loc[index, 'month_number'] + 1)))
    
    if value in list(range(1, 12, 1)):
        date_cpi_elements_pcps.append((datetime.date(year[0], next_month[0], 1) - datetime.timedelta(days = 1)).isoformat())
    else: date_cpi_elements_pcps.append(datetime.date(year[0], 12, 31).isoformat())
```

```python
# checking the 'date_cpi_elements_pcps' list
date_cpi_elements_pcps
```

```python
# assigning the newly created date list as new column
df_cpi_elements_pcps['date'] = date_cpi_elements_pcps
df_cpi_elements_pcps.head(288)
```
