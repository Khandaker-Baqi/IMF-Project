## Merging the dataframes

### Merging the CPI elements and PCPS dataframe

```python
# merging the cpi and pcps dataframe
df_cpi_elements_pcps_merged = df_cpi.merge(df_pcps, on = ['month_code'], how = 'left', indicator = True )
```

```python
# checking the column names, data types, nuber of rows and columns in the merged dataframe
df_cpi_elements_pcps_merged.info()
```

### Checking and Removing the merge flag column

```python
# checking the value of the merge column
df_cpi_elements_pcps_merged['_merge'].value_counts(dropna = False)
```

```python
# dropping the '_merge' column
df_cpi_elements_pcps_merged.drop('_merge', axis = 1, inplace = True)
df_cpi_elements_pcps_merged.info()
```

### Merging the overall CPI dataframe and PCPS dataframe

```python
df_overall_cpi_pcps_merged = df_overall_cpi.merge(df_pcps, on = ['month_code'], how = 'left', indicator = True)
df_overall_cpi_pcps_merged.info()
```

### Checking and removing the merge column

```python
# checking the merge column
df_overall_cpi_pcps_merged['_merge'].value_counts(dropna = False)
```

```python
# removing the merge column
df_overall_cpi_pcps_merged.drop('_merge', axis = 1, inplace = True)
df_overall_cpi_pcps_merged.info()
```
